```python
#二、3.
import pandas as pd

df = pd.DataFrame({
    "name": ["Tom", "Jerry", "Mike"],
    "score": [85, 92, 78],
    "gender": ["男", "女", "男"]
})

print(df[df["score"] > 80])
```

```python
#二、4.
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "name": ["Tom", "Jerry", "Mike"],
    "score": [85, 92, 78],
    "gender": ["男", "女", "男"]
})

df["grade"] = np.where(df["score"] >= 90, "优秀", "良好")

print(df)
```
